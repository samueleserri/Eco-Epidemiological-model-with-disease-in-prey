```python
from DiscreteTimeModel import DiscreteTimeModel
```

Example 1: 

b = 0.2,
c = 0.6,
d = 0.12,
k  = 0.1,
m  = 0.2,
\beta = 0.05,
K = 8, 
$r \in [0.01, 4]$.

With initial conditions:
S0 = 4, I0 = 0.5, Y0 = 0.1

```python
# instantiate the model
parameters = {
    "r": 0.001,
    "b":  0.2,
    "c": 0.6,
    "d": 0.12,
    "k": 0.1,
    "m": 0.2,
    "beta": 0.05,
    "K" : 8
}

initial_conditions = {
    "S0": 4,
    "I0": 1,
    "Y0": 2
}

model = DiscreteTimeModel(parameters)
```

$\mathcal{R}_0 = {K \beta}/{c}$

```python
print("Basic Reproduction Number R0:", model.R0)
```

```python
res = model.run(initial_conditions)
model.plot_phase_space(initial_conditions, res=res)
```

```python
# susceptible population vs parameter 
model.plot_populations_vs_parameter("r", 2, 4, initial_conditions)
```
