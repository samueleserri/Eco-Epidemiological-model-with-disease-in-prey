```python
from DiscreteTimeModel import DiscreteTimeModel
```

Example 1: 



```python

parameters = {
    "r": 0.001,
    "b":  0.2,
    "c": 0.6,
    "d": 0.12,
    "k": 0.1,
    "m": 0.2,
    "beta": 0.05,
    "K" : 8
}

initial_conditions = {
    "S0": 4,
    "I0": 0.5,
    "Y0": 0.1
}

model = DiscreteTimeModel(parameters)
```

$\mathcal{R}_0 = {K \beta}/{c}$

```python
model.print_stability_conditions_E1()
```

```python
model.r = 1
res = model.run(initial_conditions)
model.print_parameters()
model.plot_phase_space(initial_conditions, res=res)
```

```python
# susceptible population vs parameter 
model.plot_populations_vs_parameter("r", 0.01, 4, initial_conditions)
```

```python
model.plot_bifurcation("r", 0.01, 4, initial_conditions)
```
